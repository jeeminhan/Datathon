```python
import pandas as pd
import nltk
import os
from os.path import exists

nltk.download([
     "names",
     "stopwords",
     "state_union",
     "twitter_samples",
     "movie_reviews",
     "averaged_perceptron_tagger",
     "vader_lexicon",
     "punkt",
])

from nltk.sentiment import SentimentIntensityAnalyzer

song_data = pd.read_csv("songs.csv")
```

```python
def lyric_analysis(filename, year, rank, title, artist):
  result_list = []
  list2 = []
  try:
    f = open("./Datathon/Lyrics/" + filename, 'r')
  except:
    print(filename)
    pass
  content = f.read()

  sia = SentimentIntensityAnalyzer()
  result = sia.polarity_scores(content)

  result_list.append(year)
  result_list.append(rank)
  result_list.append(title)
  result_list.append(artist)
  result_list.append(result['compound'])
  result_list.append(result['pos'])
  result_list.append(result['neg'])
  result_list.append(result['neu'])
  result_list.append(content)
  list2.append(filename)
  list2.append(content)
  f.close()
  return result_list, list2
```

```python
big_df = []
big_df2 = []
counter = 1
for row in song_data.iterrows():
  title = row[1]['Title']
  artist = row[1]['Artist']
  year = row[1]['Year']
  rank = row[1]['Rank']
  filename = f"{year}-{rank}_{title}--{artist}.txt"



  # if (os.path.exists(filename) == False):
  #   print(counter, filename)
  #   counter += 1
  try:
    list1, list2 = lyric_analysis(filename, year, rank, title, artist)
    big_df.append(list1)
    big_df2.append(list2)
  except:
    pass

df = pd.DataFrame(big_df, columns = ['Year','Rank','Title','Artist','compound','pos','neg','neu','lyrics'])
lyrics_df = pd.DataFrame(big_df2, columns = ['ID', 'Lyrics'])
df

```
