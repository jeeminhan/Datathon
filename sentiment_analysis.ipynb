```python
spotify = pd.read_csv("https://raw.githubusercontent.com/jeeminhan/Datathon/main/Spotify/song_features.csv")
spotify = spotify.set_index(['Year', 'Rank'])
nlp_merge = nlp.set_index(['Year', 'Rank'])
```

```python
nlp_merge = nlp_merge.join(spotify)
nlp_merge
```

```python
df = nlp_merge
df.to_csv("Master-Song-Dataset.csv")
```

```python
df_no_lyric = df.drop(columns=['lyrics','lyric2'])
df_no_lyric
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
y = df_no_lyric.neg
X = df_no_lyric.drop(columns = ['Title', 'Artist','pos','neg','neu','compound'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
reg = LinearRegression().fit(X_train, y_train)
y_pred = reg.predict(X_test)
print(f"The root mean squared error is {mean_squared_error(y_test, y_pred)}")
```

```python
y_pred
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(9,5) , dpi = 75)
plt.plot(y_test, label = 'test')
plt.plot(y_pred, label = 'pred')
plt.title('Negative Sentiment Regression Test vs Prediction')
plt.legend()
plt.show()
```
